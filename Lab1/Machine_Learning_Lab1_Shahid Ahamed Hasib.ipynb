The following function **'lm'** will generate a random slope m and an intercept b. It will return the random m and b variables (for m random float will be generated in range of -20 to 15), (for b a random float will be generated between -20 to 15).

## Question 8
Re-make the lm function. This time, when called with an x, y it returns the optimal m and b.

You are free to either implement least-squares regression, or the gradient descent method.

def lm(x, y) -> tuple[float, float]:
    # create the optimal values for m, b
    return m, b


The following function **'lm'** implements normalized gradient descent to compute the optimal linear regression parameters for predicting y from x, and the results are returned in the original data scale.


```python
import numpy as np
x = data['lstat']
y = data['medv']

def lm(x, y, learning_rate=0.001, epochs=1000) -> tuple[float, float]:
    # Normalize data for better gradient descent performance
    x_mean, x_std = np.mean(x), np.std(x)
    y_mean, y_std = np.mean(y), np.std(y)
    x = (x - x_mean) / x_std
    y = (y - y_mean) / y_std

    # Initialize m and b
    m, b = 0, 0
    n = len(x)  # Number of data points

    # Gradient descent
    for i in range(epochs):
        # Calculate predictions
        y_pred = m * x + b

        # Calculate gradients
        dm = (-2 / n) * sum(x * (y - y_pred))
        db = (-2 / n) * sum(y - y_pred)

        # Update m and b
        m -= learning_rate * dm
        b -= learning_rate * db

    # Rescale m and b back to original data scale
    m = m * (y_std / x_std)
    b = b * y_std + y_mean - m * x_mean

    return m, b

# Calculate optimal m and b
m, b = lm(x, y)
print(m,b)
```
